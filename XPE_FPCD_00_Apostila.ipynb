```python
#@title 1.1) Estruturas de Controle (Condicional e Repetição)

## Estrutura Condicional if-elseif-else

print("Exemplo Estrutura condicional:")
variavel = 'RENDA VARIÁVEL'
if variavel =='Renda Fixa':
  print("Variável é RENDA FIXA")
elif variavel == 'Poupança':
  print('Variável é POUPANÇA')
else:
  print('Variável é RENDA VARIÁVEL')

## Estrutura de repetição - Loop for (para)

print("\nExemplo Estrutura de repetição - loop for:")
sequencia = range(0,5)
for num in sequencia:
  print(num)

## Estrutura de repetição - Loop while (enquanto)

print("\nExemplo Estrutura de repetição - loop while:")
num = 50
while num >45:
  print(num)
  num -=1       #necessário para evitar o loop infinito
```

```python
#@title 1.2) Estrutura de Dados - Listas[], Tuplas() e Dicionários{}

## Trabalhando com Listas - Sintaxe:  =[lista]
# As listas são coleções ordenadas e mutáveis de elementos.
# Podemos adicionar, remover e modificar elementos em uma lista.
print("1) Trabalhando com Listas - Sintaxe: =[lista]")

# Criando uma lista de números inteiros
numeros = [1, 2, 3, 4, 5]
print(numeros)

# Adicionando um elemento no final da lista
numeros.append(6)
print(numeros)

# Removendo um elemento específico da lista
numeros.remove(3)
print(numeros)

# Acessando elementos pelo índice
print(numeros[2])  # Saída: 4

# Modificando um elemento pelo índice
numeros[1] = 10
print(numeros)

# Tamanho da lista
print(len(numeros))  # Saída: 5

# Verificando a existência de um elemento na lista
if 10 in numeros:
    print("O número 10 está na lista.")
else:
    print("O número 10 não está na lista.")

## Trabalhando com Tuplas - Sintaxe:  =(tuplas)
print("\n2) Trabalhando com Tuplas - Sintaxe: =(tuplas)")
# As tuplas são coleções ordenadas e imutáveis de elementos.
# Uma vez criada, não podemos alterar seus elementos.

# Criando uma tupla de coordenadas
coordenadas = (10, 20)

# Acessando elementos pelo índice
x = coordenadas[0]  # x = 10
y = coordenadas[1]  # y = 20
print("A coordenada X é " , x)
print("A coordenada Y é " , y)

# Tentativa de modificação (isso resultará em erro)
# coordenadas[0] = 5  # TypeError: 'tuple' object does not support item assignment


## Trabalhando com Dicionários - Sintaxe:  ={dicionário}
print("\n3) Trabalhando com Dicionários - Sintaxe:  ={dicionário}")
# Os dicionários são coleções de pares chave-valor, onde cada valor é associado a uma chave única.
# Eles são úteis para representar estruturas de dados complexas.

# Criando um dicionário de informações de uma ação
acao = {
    "nome": "BBDC",
    "tipo": "4 - PN",
    "dy": "10.10%",
    "tag along": "80%"
}

# Acessando valores pelas chaves
print(acao["nome"])  # Saída: "BBDC"
print(acao["tipo"])  # Saída: "4 - PN"

# Modificando valores através das chaves
acao["dy"] = "6.10%"

# Adicionando um novo par chave-valor
acao["P/VP"] = 1.08

# Verificando se uma chave existe no dicionário
if "dy" in acao:
    print(f"O DY de {acao['nome']} é {acao['dy']}.")
else:
    print("Informação sobre DY não encontrada.")

print(acao)

# Removendo um elemento específico do dicionário
del acao["dy"]
print(acao)
```

**Funções embutidas em python**

Essas funções são incorporadas diretamente na linguagem e fornecem uma variedade de recursos e funcionalidades prontas para uso, sem a necessidade de importar módulos adicionais. Elas são acessíveis a partir do momento em que você começa a escrever código Python.

Você aprenderá a usar funções como print(), len(), type(), str(), int(), float(), input(), range(), list(), sorted(), sum(), max(), min(), abs(), round(), map(), filter() e zip() para manipular dados e executar operações úteis.

```python
#@title 5.1) Funções Embutidas em Python (Alguns exemplos)

# Exemplo de investimentos em stocks de tecnologia (ações)
acoes = ['AAPL', 'GOOGL', 'TSLA', 'AMZN', 'MSFT']

# 01) Função print(): Utilizada para exibir mensagens e valores na saída padrão.
print("Lista de ações:", acoes)   # Lista de ações: ['AAPL', 'GOOGL', 'TSLA', 'AMZN', 'MSFT']

# 02) Função len(): Retorna o tamanho (número de elementos) de uma sequência.
quantidade_acoes = len(acoes)
print("Quantidade de ações:", quantidade_acoes)   # Quantidade de ações: 5

# 03) Função type(): Retorna o tipo de um objeto ou variável.
tipo_dados = type(acoes)
print("Tipo de dados da lista de ações:", tipo_dados)   # Tipo de dados da lista de ações: <class 'list'>

# 04) Função str(): Converte um valor em uma representação de string.
primeira_acao = str(acoes[0])
print("Primeira ação convertida para string:", primeira_acao)   # Primeira ação convertida para string: AAPL

# 05) Função int(): Converte um valor em um número inteiro.
preco_acao = 123.45
preco_acao_inteiro = int(preco_acao)
print("Preço da ação convertido para inteiro:", preco_acao_inteiro)   # Preço da ação convertido para inteiro: 123

# 06) Função float(): Converte um valor em um número de ponto flutuante.
valor_investimento = 1000.50
valor_investimento_formatado = float("{:.2f}".format(valor_investimento))
print("Valor de investimento formatado como float:", valor_investimento_formatado)   # Valor de investimento formatado como float: 1000.5

# 07) Função input(): Lê uma entrada do usuário via teclado.
nome_investidor = input("Digite o seu nome: ")
print("Olá,", nome_investidor + "!")   # Olá, Robert Gans!

# 08) Função range(): Gera uma sequência de números em um intervalo especificado.
anos_investimento = range(1, 6)
print("Período de investimento em anos:", list(anos_investimento))   # Período de investimento em anos: [1, 2, 3, 4, 5]

# 09) Função list(): Cria uma lista a partir de uma sequência ou iterável.
retorno_investimento = {'AAPL': 0.1, 'GOOGL': 0.2, 'TSLA': 0.05, 'AMZN': 0.15, 'MSFT': 0.12}
print("Retorno de investimento por ação:", retorno_investimento)
# Retorno de investimento por ação: {'AAPL': 0.1, 'GOOGL': 0.2, 'TSLA': 0.05, 'AMZN': 0.15, 'MSFT': 0.12}
retorno_investimento_lista = list(retorno_investimento.items())
print("Retorno de investimento convertido para lista:", retorno_investimento_lista)
# Retorno de investimento convertido para lista: [('AAPL', 0.1), ('GOOGL', 0.2), ('TSLA', 0.05), ('AMZN', 0.15), ('MSFT', 0.12)]

# 10) Função sorted(): Ordena uma sequência ou iterável em ordem ascendente.
retorno_ordenado = sorted(retorno_investimento, key=retorno_investimento.get, reverse=True)
print("Ações ordenadas por retorno de investimento:", retorno_ordenado)
# Ações ordenadas por retorno de investimento: ['GOOGL', 'AMZN', 'MSFT', 'AAPL', 'TSLA']

# 11) Função sum(): Calcula a soma dos elementos de uma sequência numérica.
retorno_total = sum(retorno_investimento.values())
print("Retorno total do investimento:", retorno_total)   # Retorno total do investimento: 0.62

# 12) Função max(): Retorna o maior valor em uma sequência.
retorno_maximo = max(retorno_investimento.values())
print("Retorno máximo entre as ações:", retorno_maximo)   # Retorno máximo entre as ações: 0.2

# 13) Função min(): Retorna o menor valor em uma sequência.
retorno_minimo = min(retorno_investimento.values())
print("Retorno mínimo entre as ações:", retorno_minimo)   # Retorno mínimo entre as ações: 0.05

# 14) Função abs(): Retorna o valor absoluto de um número.
diferenca_retorno = abs(retorno_maximo - retorno_minimo)
print("Diferença entre retorno máximo e mínimo:", diferenca_retorno)   # Diferença entre retorno máximo e mínimo: 0.15000000000000002

# 15) Função round(): Arredonda um número para um determinado número de casas decimais.
retorno_arredondado = round(retorno_total, 2)
print("Retorno total arredondado:", retorno_arredondado)   # Retorno total arredondado: 0.62

# 16) Função map(): Aplica uma função a todos os elementos de uma sequência e retorna um iterador com os resultados.
precos = [100, 200, 300, 400, 500]
dobro_precos = list(map(lambda x: x * 2, precos))
print("Preços dobrados:", dobro_precos)   # Preços dobrados: [200, 400, 600, 800, 1000]

# 17) Função filter(): Filtra os elementos de uma sequência com base em uma função de teste e retorna um iterador com os elementos filtrados.
precos_altos = list(filter(lambda x: x > 250, precos))
print("Preços acima de 250:", precos_altos)   # Preços acima de 250: [300, 400, 500]

# 18) Função zip(): Combina múltiplas sequências em uma única sequência de tuplas, em que cada tupla contém elementos correspondentes das sequências originais.
acoes = ['AAPL', 'GOOGL', 'TSLA']
precos = [100, 200, 300]
acoes_precos = zip(acoes, precos)
print("Ações e preços combinados:", list(acoes_precos))   # Ações e preços combinados: [('AAPL', 100), ('GOOGL', 200), ('TSLA', 300)]

# 19) Função zip(): Mesclando 2 listas em um dicionário
acoes = ('AAPL', 'GOOGL', 'TSLA')
precos = (100, 200, 300)

dict1 = dict(zip(acoes,precos))
print("Dicionário criado a partir de 2 listas: ", dict1)
```

**Operadores em Python**

Os operadores em Python são símbolos especiais que executam operações em um ou mais operandos e retornam um resultado. Eles são usados para realizar cálculos matemáticos, comparações entre valores e manipulações de dados, além de construir expressões lógicas, aritméticas, associativas, atribuir valores a variáveis e compará-los.

Na linguagem Python temos a seguinte separação entre os diferentes tipos de operadores:


•	Operadores aritméticos


•	Operadores de atribuição


•	Operadores de comparação


•	Operadores lógicos


•	Operadores de identidade


•	Operadores de associação


```python
#@title 5.2.1) Operadores Aritméticos

n1 = 5
n2 = 2

soma = n1 + n2
subtracao = n1 - n2
multiplicacao = n1 * n2
divisao = n1 / n2
divisao_inteira = n1 // n2
modulo = n1 % n2
exponenciacao = n1 ** n2

print(soma) # 7
print(subtracao) # 3
print(multiplicacao)  # 10
print(divisao) # 2.5
print(divisao_inteira) # 2
print(modulo)  # 1
print(exponenciacao) # 25
```

```python
#@title 5.2.2) Operadores Aritméticos (tipo booleano)

print(10 + False) # Equivale a 10 + 0
print(True + 5) # Equivale a 1 + 5
print(True + True + False) # Equivale a 1 + 1 + 0
```

```python
#@title 5.2.3) Operadores Aritméticos (Regras de Precedência)

expressao = 4 + 2 * 3 - (8 / 2) ** 2

# Passo a passo:
# 1. Dentro dos parênteses: 8 / 2 = 4
# 2. Exponenciação: 4 ** 2 = 16
# 3. Multiplicação: 2 * 3 = 6
# 4. Adição e subtração da esquerda para a direita: 4 + 6 - 16 = -6

print("Portanto, o resultado da expressão é", expressao)
```

```python
#@title 5.3) Operadores de Atribuição

n1 = 5
atrib_simples = atrib_adicao = atrib_subtracao = atrib_multiplicacao = atrib_divisao = atrib_resto_div = atrib_exponenciacao = n1

atrib_simples = n1
atrib_adicao += 5
atrib_subtracao -= 2
atrib_multiplicacao *= 5
atrib_divisao /= 2
atrib_resto_div %= 2
atrib_exponenciacao **=3

print(atrib_simples) # 5
print(atrib_adicao) # 10
print(atrib_subtracao)  # 3
print(atrib_multiplicacao) # 25
print(atrib_divisao) # 2.5
print(atrib_resto_div)  # 1
print(atrib_exponenciacao) # 125
```

```python
#@title 5.4) Operadores de comparação

x = 5; y = 10; z = 20

# Operador Igual a (==)
resultado1 = x == y

# Operador Diferente de (!=)
resultado2 = y != x

# Operador Maior que (>)
resultado3 = z > y

# Operador Menor que (<)
resultado4 = x < z

# Operador Maior ou igual a (>=)
resultado5 = y >= z

# Operador Menor ou igual a (<=)
resultado6 = z <= x

# Imprimir os resultados
print(resultado1)  # False
print(resultado2)  # True
print(resultado3)  # True
print(resultado4)  # True
print(resultado5)  # False
print(resultado6)  # False
```

```python
#@title 5.5) Operadores lógicos

# Operador E lógico (and)
x = 5
resultado1 = x > 3 and x < 10  # Ambas as condições são verdadeiras

# Operador Ou lógico (or)
x = 5
resultado2 = x > 10 or x < 3  # Pelo menos uma das condições é verdadeira

# Operador Negação lógica (not)
x = 5
resultado3 = not(x < 10)  # A condição é negada

# Imprimir os resultados
print(resultado1)  # True
print(resultado2)  # False
print(resultado3)  # True
```

```python
#@title 5.6) Operadores de identidade

# Operador É (is)
x = [1, 2, 3]
y = [1, 2, 3]
z = x

resultado1 = x is y  # False, pois x e y têm identidades diferentes
resultado2 = x is z  # True, pois x e z têm a mesma identidade

# Operador Não é (is not)
resultado3 = x is not y  # True, pois x e y têm identidades diferentes

# Imprimir os resultados
print(resultado1)
print(resultado2)
print(resultado3)

# Verificando a identidade dos objetos ou variáveis
print(id(x))
print(id(y))
print(id(z))
```

```python
#@title 5.7) Operadores de Associação

renda_fixa_cliente = ["CDB","LCI","LCA","DEBENTURE","TD"]

renda_fixa_1 = "CDB"
renda_fixa_2 = "RDB"
renda_fixa_3 = "CRI"

print(renda_fixa_1 in renda_fixa_cliente) # True
print(renda_fixa_2 in renda_fixa_cliente) # False
print(renda_fixa_3 not in renda_fixa_cliente) # True
print(renda_fixa_1 not in renda_fixa_cliente) # False
```

```python
#@title 5.10.2) Strings - Trabalhando com Indices (Fatiamento de Strings) - 1) Usando Intervalo [inicial:final]

texto = "Investir em conhecimento rende sempre os melhores juros!"
print(len(texto))

substring = texto[25:40]  # Fatiando a string para obter a substring "rende sempre"
print(substring)  # Saída: 'rende sempre'
```

```python
#@title 5.10.3) Strings - Trabalhando com Indices (Fatiamento de Strings) - 2) Omitindo indice inicial ou final

texto = "Investir em conhecimento rende sempre os melhores juros!"

inicio_ate_8 = texto[:8]  # Fatiamento até o índice 8 (não incluso)
print(inicio_ate_8)  # Saída: 'Investir'

do_25_ate_fim = texto[25:]  # Fatiamento a partir do índice 25 até o fim
print(do_25_ate_fim)  # Saída: 'rende sempre os melhores juros!'
```

```python
#@title 5.14) Strings - Método find()
# Usado para encontrar a posição da primeira ocorrência de uma substring em uma string.
# Ele retorna o índice (posição) em que a substring é encontrada ou -1 caso não seja encontrada.

valores_renda = "1500, 2000, 1800, 2200"
valor_procurado = "1800"

posicao = valores_renda.find(valor_procurado)

mensagem = (f"O valor de renda {valor_procurado} foi encontrado na posição {posicao}."
            if posicao != -1
            else f"O valor de renda {valor_procurado} não foi encontrado na string.")

print(mensagem)
```

```python
#@title 5.25) Trabalhando com LOOPS em Python (Loop while) - Exemplo 1
# 1. Verificação de Rentabilidade com Loop While
# Vamos agora criar um exemplo usando um loop while para verificar a rentabilidade de um investimento.

# Dado de investimento
investimento_inicial = 1500
valor_atual = 2000

# Cálculo da rentabilidade
rentabilidade = (valor_atual - investimento_inicial) / investimento_inicial * 100

# Verificação com loop while
i = 1
while rentabilidade < 20:
    valor_atual *= 1.05  # Incrementa 5% do valor atual
    rentabilidade = (valor_atual - investimento_inicial) / investimento_inicial * 100
    i += 1

print(f"O investimento atingiu rentabilidade de 20% após {i} anos.")
```

```python
#@title 5.26) Trabalhando com LOOPS em Python (Loop while) - Exemplo 2
# 2. Atingimento de objetivo de investimento com Loop While
# Vamos agora criar um exemplo usando um loop while para atingir um objetivo financeiro em um investimento.

# Rendimento mensal de uma ação (exemplo fictício)
rendimento_mensal = 3.2

# Capital inicial e objetivo desejado
capital_inicial = 10000
objetivo = 15000 # Objetivo de atingir um valor de 15000 após investir 10000 com rendimento mensal de 3.2%

# Simulação da estratégia com while e for
capital_atual = capital_inicial
mes = 0

while capital_atual < objetivo:
    mes += 1
    capital_atual *= (1 + rendimento_mensal / 100)
    print(f"Após o mês {mes}, o capital é de R${capital_atual:.2f}.")

print(f"Objetivo de R${objetivo:.2f} atingido após {mes} meses.")
```

```python
#@title 5.26a) Trabalhando com LOOPS em Python (Loop while) - Exemplo 2 (uso de continue e break)
# Uso de 'continue' e 'break' em loops de repetição
# Mesmo exemplo anterior com novas condições

# Rendimento mensal de uma ação (exemplo fictício)
rendimento_mensal = 3.2

# Capital inicial e objetivo desejado
capital_inicial = 10000
objetivo = 15000 # Objetivo de atingir um valor de 15000 após investir 10000 com rendimento mensal de 3.2%

# Simulação da estratégia com while e for
capital_atual = capital_inicial
mes = 0

while capital_atual < objetivo:
    mes += 1
    # Introdução do uso do continue
    if mes == 3:
        print("Pulando o mês 3.")
        continue
    capital_atual *= (1 + rendimento_mensal / 100)
    print(f"Após o mês {mes}, o capital é de R${capital_atual:.2f}.")
    # Introdução do uso do break
    if capital_atual > objetivo * 0.8:
        print("Objetivo multiplicado por 0.8 atingido. Encerrando a simulação.")
        break
```

```python
#@title 7.4) Biblioteca Numpy - Funções Universais (UFuncs)
# funções matemáticas e lógicas
# Essas funções são vetorizadas, o que significa que elas operam em todos os elementos de um array de uma só vez.

import numpy as np

# Função de exponenciação > np.exp()
arr = np.array([1, 2, 3, 4, 5])
result = np.exp(arr)
print(result)  # Saída: [ 2.71828183  7.3890561  20.08553692 54.59815003 148.4131591 ]

# Função de Seno > np.sin()
arr = np.array([0, np.pi/2, np.pi])
result = np.sin(arr)
print(result)  # Saída: [0.         1.         1.2246468e-16]

# Função de Cosseno > np.cos()
arr = np.array([0, np.pi/2, np.pi])
result = np.cos(arr)
print(result)  # Saída: [ 1.000000e+00  6.123234e-17 -1.000000e+00]

# Função de Raiz Quadrada > np.sqrt()
arr = np.array([4, 9, 16])
result = np.sqrt(arr)
print(result)  # Saída: [2. 3. 4.]

# Função de Logaritmo > np.log()
arr = np.array([1, np.e, np.e**2])
result = np.log(arr)
print(result)  # Saída: [0.         1.         2.        ]

# Função de soma > np.sum()
arr = np.array([1, 2, 3, 4, 5])
result = np.sum(arr)
print(result)  # Saída: 15

# Função de elemento máximo > np.max()
arr = np.array([1, 2, 3, 4, 5])
result = np.max(arr)
print(result)  # Saída: 5

# Função de elemento mínimo > np.min()
arr = np.array([1, 2, 3, 4, 5])
result = np.min(arr)
print(result)  # Saída: 1
```

```python
#@title 9.6) Biblioteca Matplotlib - Gráfico de Áreas [ plt.fill_between() ou plt.stackplot() ]:

import matplotlib.pyplot as plt
import random

# Dados de exemplo - Preços hipotéticos das ações de duas empresas ao longo do tempo
tempo = list(range(1, 11))  # 10 pontos de dados no tempo
empresa_a_precos = [random.uniform(100, 200) for _ in tempo]  # Preços da Empresa A
empresa_b_precos = [random.uniform(80, 180) for _ in tempo]   # Preços da Empresa B

# Calcular a soma acumulada dos preços das ações para cada empresa utilizando list comprehension e a função sum()
empresa_a_acumulado = [sum(empresa_a_precos[:i+1]) for i in range(len(empresa_a_precos))]
empresa_b_acumulado = [sum(empresa_b_precos[:i+1]) for i in range(len(empresa_b_precos))]

# Plotar o Gráfico de Áreas
plt.figure(figsize=(8, 6))  # define o tamanho do gráfico

plt.stackplot(tempo, empresa_a_acumulado, empresa_b_acumulado, labels=['Empresa A', 'Empresa B'], alpha=0.7)
# plt.stackplot(x(tempo), y(lista valores para áreas), labels = rótulos às legendas, alpha = define a transparência das áreas )

plt.xlabel('Tempo') # Rótulo eixo X
plt.ylabel('Soma Acumulada do Preço da Ação') # Rótulo eixo Y
plt.title('Soma Acumulada dos Preços das Ações de Empresa A e Empresa B ao Longo do Tempo') # Título do Gráfico
plt.legend(loc='upper left')  # é usado para mostrar as legendas no canto superior esquerdo do gráfico.
plt.grid(True)  # adiciona linhas de grade ao gráfico para facilitar a leitura.
plt.tight_layout()  # é usado para evitar que as legendas se sobreponham ao gráfico.
plt.show()  # exibe o gráfico
```

```python
#@title 9.7) Biblioteca Matplotlib - Gráfico de Histograma [ plt.hist() ]:

import matplotlib.pyplot as plt
import random

# Dados de exemplo - Retornos diários hipotéticos de uma ação
# Criação de uma lista (retornos_diarios) com 100 valores hipotéticos de uma ação
num_dias = 100
retornos_diarios = [random.normalvariate(0.01, 0.02) for _ in range(num_dias)]
# Os valores são gerados aleatoriamente utilizando a função random.normalvariate(), que gera números aleatórios
# de acordo com uma distribuição normal com média de 0.01 e desvio padrão de 0.02.

# Plotar o Histograma
plt.figure(figsize=(8, 6)) # define o tamanho do gráfico

plt.hist(retornos_diarios, bins=20, density=True, edgecolor='black', alpha=0.7)
# <lista dos retornos diários> dados a serem plotados
# <bins> define o número de bins (intervalos) no histograma, que influencia na resolução da distribuição.
# <density> é usado para normalizar o histograma, para que a área total sob o histograma seja igual a 1 (densidade).
# <edgecolor> define a cor das bordas dos retângulos do histograma.
# <alpha> define a transparência das barras.

plt.xlabel('Retornos Diários')  # Rótulo eixo X
plt.ylabel('Densidade') # Rótulo eixo Y
plt.title('Histograma dos Retornos Diários de uma Ação')  # Título do Gráfico
plt.grid(True)  # adiciona linhas de grade ao gráfico para facilitar a leitura.
plt.tight_layout()  # é usado para evitar que as legendas se sobreponham ao gráfico.
plt.show()  # exibe o gráfico
```

```python
#@title 9.8) Biblioteca Matplotlib - Gráfico de Boxplotde [ plt.boxplot() ]:

import matplotlib.pyplot as plt
import random

# Dados de exemplo - Retornos diários hipotéticos de duas ações (gerados aleatoriamente com a função random.normalvariate())
# Criação de duas listas (retornos_acao1 e retornos_acao2) com 100 valores hipotéticos de cada ação
num_dias = 100
acoes = ['Ação 1', 'Ação 2']

retornos_acao1 = [random.normalvariate(0.01, 0.02) for _ in range(num_dias)]  # Média: 0.01, Desvio padrão: 0.02
retornos_acao2 = [random.normalvariate(0.02, 0.03) for _ in range(num_dias)]  # Média: 0.02, Desvio padrão: 0.03

# Plotar o Gráfico de Boxplot
plt.figure(figsize=(8, 6))  # define o tamanho do gráfico

plt.boxplot([retornos_acao1, retornos_acao2], labels=acoes, vert=False, patch_artist=True)
# <listas dos retornos diários das 2 ações> lista de dados que queremos comparar
# <labels> é usado para adicionar rótulos aos boxplots (nomes das ações).
# <vert> é usado para exibir os boxplots horizontalmente.
# <patch_artist> é usado para preencher os boxplots com cores.

plt.xlabel('Retornos Diários')  # Rótulo eixo X
plt.title('Boxplot dos Retornos Diários de Duas Ações') # Título do Gráfico
plt.grid(True)  # adiciona linhas de grade ao gráfico para facilitar a leitura.
plt.tight_layout()  # é usado para evitar que as legendas se sobreponham ao gráfico.
plt.show()  # exibe o gráfico
```

```python
#@title 10.3.1) Exemplo prático com Scikit_learn - Modelo de Regressão linear (1)

import numpy as np
from sklearn.linear_model import LinearRegression

# Dados de entrada
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

# Treinar o modelo de regressão linear
reg = LinearRegression().fit(X, y)

# Realizar a previsão para novos valores
novos_valores = np.array([[6], [7]])
previsao = reg.predict(novos_valores)

# Imprimir a previsão
print("O resultado numérico desta Regressão Linear é:")
print() # Este print sem argumento gera uma linha em branco
print(previsao)
print() # Este print sem argumento gera uma linha em branco

# Vamos ver o exemplo através de um gráfico para melhor entendimento

#Importando a biblioteca necessária
import matplotlib.pyplot as plt

print("O resultado gráfico desta Regressão Linear é:")
print() # Este print sem argumento gera uma linha em branco

# Plotar o gráfico de dispersão e a linha de regressão
plt.scatter(X, y, color='black')
plt.plot(X, reg.predict(X), color='blue', linewidth=3, label='Regressão Linear')

# Plotar a previsão como uma nova linha
plt.plot(novos_valores, previsao, color='green', linestyle='--', linewidth=3, label='Previsão')

# Configurar as legendas do gráfico
plt.title('Regressão linear simples')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

A saída da regressão linear é uma equação matemática que modela a relação entre as variáveis de entrada e de saída. Para cada valor de entrada, a equação fornece uma previsão para o valor de saída correspondente. No exemplo de código Python que apresentei anteriormente, a saída é uma previsão para novos valores de entrada.

```python
#@title 10.3.2) Exemplo prático com Scikit_learn - Modelo de Regressão linear (2)

# Importando a função make_regression do módulo sklearn.datasets.
# Essa função é utilizada para criar dados sintéticos para problemas de regressão.
from sklearn.datasets import make_regression

# Gerando uma massa de dados:
x, y = make_regression(n_samples=200, n_features=1, noise=30)
# Estamos gerando 200 amostras, cada uma com uma única feature (1 característica) e adicionando ruído de valor 30.
# O conjunto de dados gerado é armazenado nas variáveis x e y, onde x contém as features e y contém os valores alvo (rótulos).

# Importando a biblioteca de gráficos
import matplotlib.pyplot as plt

# Mostrando no gráfico:
# Criando um gráfico de dispersão (scatter). O gráfico mostra os pontos de dados do conjunto x em relação aos valores alvo y.
plt.scatter(x,y)
print("Modelo de Regressão Linear\n")
print("ETAPA 01: Gráfico de Dispersão com os dados gerados aleatóriamente para o nosso modelo de Regressão\n")
plt.show()  # Exibe o gráfico

# Aqui, estamos importando a classe LinearRegression do módulo sklearn.linear_model.
# Essa classe é usada para criar um modelo de regressão linear.
from sklearn.linear_model import LinearRegression
# Criação do modelo
modelo = LinearRegression()

# Aqui, estamos ajustando o modelo aos dados x e y.
# Ou seja, estamos treinando o modelo com os dados de entrada x e os valores alvo y.
modelo.fit(x,y)

# Nesta linha, estamos utilizando o modelo treinado para fazer previsões com base nos dados de entrada x.
# O resultado é uma previsão para cada ponto em x.
modelo.predict(x)

# Nesta linha, estamos novamente criando um gráfico de dispersão com os pontos de dados x em relação aos valores alvo y.
plt.scatter(x,y)
# Aqui, estamos usando a função plot do matplotlib.pyplot para traçar uma linha que representa as previsões feitas pelo modelo em relação aos dados x.
# A linha é desenhada em vermelho (color='red') com uma espessura de linha de 3 pixels (linewidth=3).
plt.plot(x, modelo.predict(x), color='red', linewidth=3)
print("ETAPA 02: Gráfico de Dispersão com a reta gerada pelo nosso modelo de regressão baseado nos pontos\n")
plt.show()  # Exibe o gráfico
# Esta linha exibe o gráfico contendo os pontos de dados originais x em relação aos valores alvo y e
# a linha de previsão em vermelho feita pelo modelo de regressão linear.
```

```python
#@title 10.3.3) Exemplo prático com Scikit_learn - Clusterização (1)

# Este código é um exemplo simples de como usar a biblioteca Scikit-learn em Python para executar o algoritmo de clusterização K-means em um conjunto de dados.

# Primeiramente, o código importa as bibliotecas necessárias:
# NumPy para trabalhar com matrizes e o KMeans do Scikit-learn para executar o algoritmo de clusterização.

# Importando bibliotecas necessárias
import numpy as np
from sklearn.cluster import KMeans

# Depois, um conjunto de dados é definido como a variável X, que é uma matriz NumPy de seis pontos em um espaço bidimensional.

# Dados de entrada
X = np.array([[1, 2], [1, 4], [1, 0], [4, 2], [4, 4], [4, 0]])

# O próximo passo é treinar o modelo de clusterização usando a função KMeans do Scikit-learn, que recebe como argumento o número de clusters que deseja-se identificar.
# Neste caso, o número de clusters é definido como 2.

# Treinar o modelo de clusterização
kmeans = KMeans(n_clusters=2).fit(X)

# Obter os rótulos de cluster para cada objeto
rotulos = kmeans.labels_

# Imprimir os rótulos de cluster
print("O resultado numérico desta Análise de Cluster é:")
print() # Este print sem argumento gera uma linha em branco
print(rotulos)
print() # Este print sem argumento gera uma linha em branco

# Vamos ver o exemplo através de um gráfico para melhor entendimento

#Importando biblioteca necessária para a plotagem do gráfico
import matplotlib.pyplot as plt

print("O resultado gráfico desta Análise de Cluster é:")
print() # Este print sem argumento gera uma linha em branco

# Criar uma figura e um eixo
fig, ax = plt.subplots()

# Adicionar os pontos ao eixo
ax.scatter(X[:,0], X[:,1], c=rotulos)

# Exibir o gráfico
plt.title('Análise de Cluster')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

A saída da análise de cluster é um conjunto de rótulos de cluster, onde cada rótulo identifica a que grupo pertence cada objeto. Os grupos são formados com base em semelhanças entre os objetos, onde objetos semelhantes são atribuídos ao mesmo grupo. No exemplo de código Python que apresentei anteriormente, a saída são os rótulos dos grupos (0 ou 1) para cada objeto de entrada.

```python
#@title 10.3.4) Exemplo prático com Scikit_learn - Clusterização (2)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Importa a classe KMeans do módulo cluster da biblioteca scikit-learn.
# Essa classe é usada para realizar a clusterização com o algoritmo k-means.
from sklearn.cluster import KMeans
# Importa a classe StandardScaler do módulo preprocessing da biblioteca scikit-learn.
# Essa classe é usada para padronizar os dados antes da clusterização.
from sklearn.preprocessing import StandardScaler

# Cria um dicionário com dados fictícios de indicadores fundamentalistas para seis empresas diferentes.
dados = {
    'Empresa': ['Empresa A', 'Empresa B', 'Empresa C', 'Empresa D', 'Empresa E', 'Empresa F'],
    'Receita (milhões)': [100, 200, 150, 300, 250, 180],
    'Lucro Líquido (milhões)': [20, 30, 25, 40, 35, 28],
    'Margem de Lucro (%)': [20, 15, 16.67, 13.33, 14, 15.56],
}

# Converter os dados em um DataFrame
df = pd.DataFrame(dados)

# Seleciona apenas os indicadores fundamentalistas (Receita, Lucro Líquido e Margem de Lucro) para a clusterização.
# A coluna "Empresa" é removida porque não é relevante para a análise de padrões.
X = df.drop('Empresa', axis=1)

# Padronizar os dados para que tenham média zero e desvio padrão igual a um
# Cria uma instância do StandardScaler para padronizar os dados.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Realizar a clusterização com k-means (vamos assumir 2 clusters)
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Adicionar os rótulos dos clusters aos dados originais
df['Cluster'] = clusters

# Visualizar os resultados
print(df)

# Plotar os clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200, c='red')
plt.xlabel('Receita (padronizado)')
plt.ylabel('Lucro Líquido (padronizado)')
plt.title('Clusters de Empresas')
plt.show()
```

Neste exemplo, criamos um conjunto fictício de dados com indicadores fundamentalistas para seis empresas. Em seguida, selecionamos apenas os indicadores fundamentalistas para a clusterização. Usamos o StandardScaler para padronizar os dados, o que é uma etapa importante para garantir que as variáveis estejam em escalas comparáveis.

Aplicamos o algoritmo k-means com 2 clusters (você pode ajustar o número de clusters conforme necessário). Os rótulos dos clusters são adicionados de volta ao DataFrame original e, em seguida, plotamos os resultados usando um gráfico de dispersão, onde as cores representam os diferentes clusters.
